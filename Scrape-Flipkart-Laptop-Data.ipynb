```python
# Import library
import selenium
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
import requests
```

## Flipkart

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/')
driver.maximize_window()
```

```python
destination = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
destination.send_keys('laptop')
```

```python
search = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
search.click()
```

```python
product_url=[]
laptop_name=[]
price=[]
rating=[]
detail=[]
```

```python
# scrape all product_url
for i in range(5):
    print("Scraping page", i+1)
    titles=driver.find_elements(By.XPATH, '//a[@class="CGtC98"]')
    for i in titles:
        product_url.append(i.get_attribute('href'))

    next_page = driver.find_element(By.XPATH, '//a[@class="_9QVEpD"]')
    next_page.click()
    time.sleep(3)
```

```python
for i in enumerate(product_url):
    print(i)
```

```python
i[0]
```

```python
product_url = product_url[0:20]
```

```python
laptop_name  =[]
price = []
rating = []
detail = []
```

```python
for i in product_url:
    driver.get(i)
    time.sleep(5)

    read_more = driver.find_element(By.XPATH, '//button[@class="QqFHMw _4FgsLt"]')
    read_more.click()

    try:
        detail_elements = driver.find_elements(By.XPATH, '//li[@class="HPETK2"]')
        detail_text = '|'.join([element.text for element in detail_elements])   # Cancatenate all texts
        detail.append(detail_text)  # Append as a single string
    except NoSuchElementException:
        detail.append('-')  # Handle the case where the elements are not found

    try:
        laptop_tag = driver.find_element(By.XPATH, '//span[@class="VU-ZEz"]')
        laptop_name.append(laptop_tag.text)
    except NoSuchElementException:
        laptop_name.append('-')

    try:
        rs = driver.find_element(By.XPATH, '//div[@class="Nx9bqj CxhGGd"]')
        price.append(rs.text)
    except NoSuchElementException:
        price.append()

    try:
        rates = driver.find_element(By.XPATH, '//div[@class="XQDdHH"]')
        rating.append(rates.text)
    except NoSuchElementException:
        rating.append('-')
```

```python
flipkart = pd.DataFrame({'Laptop Name':laptop_name, 'Price':price, 'Rating':rating, 'Detail':detail})
flipkart
```
